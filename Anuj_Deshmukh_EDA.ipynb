```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv("Customers.csv")
```

```python
customers.head()
```

```python
customers.info()
```

```python
products = pd.read_csv("Products.csv")
```

```python
products.head()
```

```python
products.info()
```

```python
transactions = pd.read_csv("Transactions.csv")
```

```python
transactions.head()
```

```python
transactions.info()
```

```python
merged_data = transactions.merge(customers, on="CustomerID").merge(products, on="ProductID")
merged_data
```

```python
merged_data.info()
```

```python
merged_data.drop_duplicates(inplace=True)
merged_data.dropna(inplace=True)
```

```python
merged_data.describe()
```

```python
sns.countplot(data=merged_data, x="Region")
plt.title("Transactions by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
top_categories = merged_data.groupby("Category")["Quantity"].sum().sort_values(ascending=False)
top_categories.plot(kind="bar", color="skyblue")
plt.title("Top-Selling Product Categories")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
merged_data["TransactionDate"] = pd.to_datetime(merged_data["TransactionDate"])
sales_by_date = merged_data.groupby("TransactionDate")["TotalValue"].sum()
sales_by_date.plot(kind="line", figsize=(10, 5))
plt.title("Sales Trends Over Time")
plt.ylabel("Total Sales Value")
plt.xlabel("Date")
plt.show()
```

```python
top_customers = merged_data.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False).head(10)
print("Top 10 Customers by Total Value of Transactions:")
print(top_customers)
```

```python
region_revenue = merged_data.groupby("Region")["TotalValue"].sum()
region_revenue.plot(kind="pie", autopct='%1.1f%%', startangle=90, figsize=(7, 7))
plt.title("Revenue Contribution by Region")
plt.ylabel("")
plt.show()
```

```python
customers["SignupDate"] = pd.to_datetime(customers["SignupDate"])
merged_data["TransactionDate"] = pd.to_datetime(merged_data["TransactionDate"])
```

```python
customer_first_transaction = merged_data.groupby("CustomerID")["TransactionDate"].min()
customers["DaysToFirstTransaction"] = (customer_first_transaction - customers.set_index("CustomerID")["SignupDate"]).dt.days
```

```python
avg_days_to_first = customers["DaysToFirstTransaction"].mean()
print(f"Average days between signup and first transaction: {avg_days_to_first}")
```

```python
customers["DaysToFirstTransaction"].hist(bins=20, color="orange")
plt.title("Distribution of Days to First Transaction")
plt.xlabel("Days")
plt.ylabel("Number of Customers")
plt.show()
```

```python
customer_total_value = merged_data.groupby("CustomerID")["TotalValue"].sum()
customer_segments = pd.cut(customer_total_value, bins=[0, 100, 500, 1000, float("inf")], 
                           labels=["Low Value", "Mid Value", "High Value", "Very High Value"])

customer_segments.value_counts().plot(kind="bar", color="purple")
plt.title("Customer Segmentation by Transaction Value")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
transaction_frequency = merged_data.groupby("CustomerID")["TransactionID"].count()

avg_frequency = transaction_frequency.mean()
print(f"Average transaction frequency per customer: {avg_frequency}")

transaction_frequency.hist(bins=20, color="green")
plt.title("Distribution of Transaction Frequency")
plt.xlabel("Number of Transactions")
plt.ylabel("Number of Customers")
plt.show()
```
